```python
import docx

def split_docx(input_file, max_paragraphs):
    # Open the input docx file
    doc = docx.Document(input_file)

    # Split the document into chunks based on the maximum paragraph count
    chunks = []
    current_chunk = docx.Document()
    for paragraph in doc.paragraphs:
        current_chunk.add_paragraph(paragraph.text)
        if len(current_chunk.paragraphs) == max_paragraphs:
            chunks.append(current_chunk)
            current_chunk = docx.Document()
    if current_chunk.paragraphs:
        chunks.append(current_chunk)

    return chunks


input_file = 'D:\Article review DH sample.docx'
max_paragraphs = 15
chunks = split_docx(input_file, max_paragraphs)
print(f'Successfully split {input_file} into {len(chunks)} chunks:')
for i, chunk in enumerate(chunks):
    print(f'Chunk {i+1} of {len(chunks)}:')
    for paragraph in chunk.paragraphs:
        print(paragraph.text)
    print('-' * 20)
```
